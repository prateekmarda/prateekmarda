Write a function that reverses a list without using the built-in reverse() method.

```python
def revlist(my_list):
  reverse = my_list[::-1]
  print(reverse)
```

```python
revlist([1,2,3])
```

```python
def parity(my_list):
  for i in my_list:
    if i % 2 == 0:
      print("Even")
    else:
      print("Odd")
```

```python
sub(3,2)
```

```python
parity([1,2,3,4,5])
```

```python
def Avg(**kwargs):
  avg = round(sum(kwargs.values())/len(kwargs.values()),2)
  return avg
```

```python
Avg(num1 = 1,num2 = 2, num3 = 25)
```

Write a function that finds the second largest number in a list.

my_list = [2, 5, 9, 3, 4, 8, 8, 7]

```python
def Second_largest(my_list):
  num = list(set(my_list))
  num.sort()
  return num[-2]
```

```python
my_list = [2, 5, 9, 3, 4, 8, 8, 7, 9,10]
```

```python
Second_largest(my_list)
```
