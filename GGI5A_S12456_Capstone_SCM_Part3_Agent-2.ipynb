```python
from langchain_core.documents import Document
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

documents = []
for element in elements:
    metadata = element.metadata.to_dict()
    documents.append(Document(page_content=element.text, metadata=metadata))

# Using the sentence-transformer model compatible with the installed libraries
db = FAISS.from_documents(documents, HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
import torch
from transformers import PreTrainedModel, AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

base_model_id = "meta-llama/Llama-3.1-8B-Instruct"
adapter_path = "prateekmarda/lora_model" # Local folder or HF Hub ID

# 1. Load the Tokenizer
tokenizer = AutoTokenizer.from_pretrained(base_model_id)

# 2. Load the Base Model
# Note: device_map="auto" handles multi-GPU/CPU placement automatically
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)

# 3. Load the LoRA Adapter
model = PeftModel.from_pretrained(base_model, adapter_path)
```

```python
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
from transformers import pipeline
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda

terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.2,
    do_sample=True,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=100,
    eos_token_id=terminators,
)

# Correctly instantiate the LangChain HuggingFacePipeline with the transformers pipeline
llm_for_chain = HuggingFacePipeline(pipeline=text_generation_pipeline)

def format_docs_with_citations(docs):
    formatted_string = ""
    for i, doc in enumerate(docs):
        source_info = ""
        if 'filename' in doc.metadata and 'page_number' in doc.metadata:
            source_info = f" (Source: {doc.metadata['filename']}, Page: {doc.metadata['page_number']})"
        elif 'filename' in doc.metadata:
            source_info = f" (Source: {doc.metadata['filename']})"
        elif 'page_number' in doc.metadata:
            source_info = f" (Page: {doc.metadata['page_number']})"

        formatted_string += f"[{i+1}] {doc.page_content}{source_info}\n\n"
    return formatted_string

prompt_template = """
<|start_header_id|>user<|end_header_id|>
You are an assistant for answering questions about Supply Chain.
You are given the extracted parts of a supply chain knowledge document and a question. Provide a conversational answer.
If you don't know the answer, just say "I do not know." Don't make up an answer.
When providing an answer, cite your sources using the numerical references provided
in the context (e.g., [1], [2]) and provide links if possible for your citations.

Question: {question}
Context: {context}<|eot_id|><|start_header_id|>assistant<|end_header_id|>
"""

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_template,
)

# Use the correctly instantiated LangChain LLM in the chain
llm_chain = prompt | llm_for_chain | StrOutputParser()
rag_chain = (
    {"context": retriever | RunnableLambda(format_docs_with_citations), "question": RunnablePassthrough()}
    | llm_chain
)
```

## Build RAG pipeline

```python
from langgraph.graph import StateGraph , START, END
```

```python
from typing import List, TypedDict
from langchain_core.documents import Document

class AgentState(TypedDict):
    query: str
    documents: List[Document]
    answer: str
    # Add missing keys to AgentState
    context: str
    prompt: str
    response: str
    source: str
    is_relevant: str
    iteration_count: int # Changed type to int for consistency
```

```python
def get_llm_response(state: AgentState):
    """Function to get response from the fine-tuned LLM, acting as a graph node."""
    print("--- GENERATING RESPONSE ---")
    input_prompt = state["prompt"] # Extract the prompt string from the state
    llm_output = llm_for_chain.invoke(input_prompt)
    state["response"] = llm_output
    return state
```

```python
def build_prompt(state: AgentState):
    """Builds the prompt for the LLM based on query and context."""
    print("--- BUILDING PROMPT ---")
    query = state["query"]
    context = state["context"]
    # Reusing the existing prompt_template for consistency (defined in _az_tYowOWxo)
    formatted_prompt = prompt_template.format(question=query, context=context)
    state["prompt"] = formatted_prompt
    return state
```

```python
from typing import Literal

# Helper function for internal LLM calls within nodes like router and relevance check
def _call_llm_for_decision(prompt_text: str) -> str:
    return llm_for_chain.invoke(prompt_text)

def retrieve_context_q_n_a(state):
  """Retreive top info from ChromaDB Collection1 based on user query"""
  print("--RETRIEVING CONTEXT--")
  query = state['query']
  results = collection1.query(query_texts=[query], n_results= 4)
  context = "\n".join(results["documents"][0])
  state["context"] = context
  state["source"] = "supplychain_info"
  print(context)

  return state

def tavily_web_search(state):
  """Perform web search using the Tavily API"""
  print("--- PERFORMING TAVILY WEBSEARCH ---")
  tavily_search = TavilySearch(topic = "general", max_results = 1)
  query = state['query']
  result = tavily_search.invoke({'query': query})
  state["answer"] = result['results'][0]['content']

  return state

def router(state: AgentState) -> Literal[
    "Retreive from document", "Web Search"
]:

  """ Agentic router decides which retreival method to use:"""
  print(f"State received by router: {state}") # Added for debugging
  query = state["query"]

  # A light weight decision LLM - you can replace this with GPT-4o-mini etc..

  decision_prompt =  f"""

  you are a router that decides which retreival path to take. Based on user query, decide where to look for information.

  Options:
  - Retreive from document: If it is about general supply chain information
  - Web Search : If the information is about recent supply chain news, disruptions or external data

  Query: "{query}"

  Respond ONLY with one of: Retreive from document, Web Search

  """

  router_raw_output = _call_llm_for_decision(decision_prompt).strip() # Use helper function
  print(f"--ROUTER RAW OUTPUT: {router_raw_output}--") # Debugging raw output

  router_decision = ""
  if "Retrieve from document" in router_raw_output:
      router_decision = "Retrieve from document"
  elif "Web Search" in router_raw_output:
      router_decision = "Web Search"
  else:
      # Fallback if LLM doesn't respond as expected
      print("--ROUTER LLM FAILED TO PARSE, DEFAULTING TO RETRIEVE FROM DOCUMENT--")
      router_decision = "Retrieve from document" # Corrected spelling

  print(f"--ROUTER DECISION: {router_decision}--")

  state["source"] = router_decision

  return state

def route_decision(state: AgentState) -> str:
    return state["source"]

def check_relevance(state: AgentState):
  """Determine whether the retrieved context is relevant or not"""
  print("--- CONTEXT RELEVANCE CHECK ---")
  query = state["query"]
  context = state["context"]

  relevance_prompt = f"""
    you are a relevance checker, check the below context if it is relevant to the user query or not.
    ####
    Context:
    {context}
    ####
    User Query: {query}

    Options:
    - Yes: If the context is relevant to the query
    - No: If the context is not relevant to the query

    Please answer with only 'Yes' or 'No'
    """

  relevance_raw_output = _call_llm_for_decision(relevance_prompt).strip() # Use helper function
  print(f"--RELEVANCE RAW OUTPUT: {relevance_raw_output}--") # Debugging raw output

  relevance_check = ""
  if "Yes" in relevance_raw_output:
      relevance_check = "Yes"
  elif "No" in relevance_raw_output:
      relevance_check = "No"
  else:
      print("--RELEVANCE LLM FAILED TO PARSE, DEFAULTING TO 'No'--")
      relevance_check = "No" # Default to 'No' for safety

  print(f"--RELEVANCE DECISION:{relevance_check}--")
  state["is_relevant"] = relevance_check

  return state

def relevance_decision(state: AgentState) -> str:
  iteration_count = state.get("iteration_count",0)
  iteration_count += 1
  state["iteration_count"] = iteration_count
  # Limiting to max 3 tries
  if iteration_count >= 3:
    print("--MAX ITERATIONS REACHED,FORCING 'Yes'--")
    state["is_relevant"] = "Yes"
  return state["is_relevant"]
```

## Define State Graph

```python
from typing import TypedDict

# Using AgentState defined previously for consistency
# class GraphState(TypedDict):
#   query: str
#   context: str
#   prompt: str
#   response: str
#   source: str
#   is_relevant: str
#   iteration_count: str

workflow = StateGraph(AgentState)

workflow.add_node("Router",router)
workflow.add_node("Retrieve from document",retrieve_context_q_n_a)
workflow.add_node("Web Search",tavily_web_search)
workflow.add_node("Check Context Relevance",check_relevance)
workflow.add_node("Augment",build_prompt)
workflow.add_node("Generate Response",get_llm_response)


workflow.add_edge(START, "Router")
workflow.add_conditional_edges(
    "Router",
    route_decision,
    {
        "Retrieve from document": "Retrieve from document",
        "Web Search": "Web Search",
    }
)
workflow.add_edge("Retrieve from document", "Check Context Relevance")
workflow.add_edge("Web Search", "Check Context Relevance")
workflow.add_conditional_edges(
    "Check Context Relevance",
    relevance_decision,
    {
        "Yes": "Augment",
        "No": "Web Search",
    }
)
workflow.add_edge("Augment", "Generate Response")
workflow.add_edge("Generate Response",END)

agentic_rag = workflow.compile()
```
